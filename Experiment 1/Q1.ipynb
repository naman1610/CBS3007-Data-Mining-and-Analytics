```python
import pandas as pd

# Read the CSV file
df = pd.read_csv('employee.csv')

# Get the dimensions of the dataset
dimensions = df.shape

# Get the structure of the dataset
structure = df.dtypes

# Get the attribute names
attribute_names = df.columns.tolist()

# Get the attribute values
attribute_values = df.values.tolist()

# Print the results
print("Dimensions:", dimensions)
print("Structure:", structure)
print("Attribute Names:", attribute_names)
print("Attribute Values:", attribute_values)
```

```python
# Display the first 5 records
print("First 5 Records:")
print(df.head(5))

# Display the last 5 records
print("Last 5 Records:")
print(df.tail(5))

# Display the Name, Designation, and Salary of the first 10 records
print("Name, Designation, Salary of First 10 Records:")
print(df[['Name', 'Designation', 'Salary']].head(10))

# Display the Name of all records
print("Name of All Records:")
print(df['Name'])

# Display all records
print("All Records:")
print(df)
```

```python
# a) Mean, median and mode of the numerical variables
print("Mean, median and mode of the numerical variables:")
mean_values = df[['Salary', 'Experience']].mean()
median_values = df[['Salary', 'Experience']].median()
mode_values = df[['Salary', 'Experience']].mode().iloc[0]

print("Mean of the numerical variables:")
print(mean_values)
print("\nMedian of the numerical variables:")
print(median_values)
print("\nMode of the numerical variables:")
print(mode_values)

# b) Variance and Covariance
variance_values = df[['Salary', 'Experience']].var()
covariance_values = df[['Salary', 'Experience']].cov()

print("\nVariance of the numerical variables:")
print(variance_values)
print("\nCovariance of the numerical variables:")
print(covariance_values)

# c) Correlation of salary to experience
correlation_salary_experience = df['Salary'].corr(df['Experience'])

print("\nCorrelation of Salary to Experience:")
print(correlation_salary_experience)
```

```python
import matplotlib.pyplot as plt

# A) Pie chart on designation
designation_counts = df['Designation'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(designation_counts, labels=designation_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Pie Chart of Designation')
plt.show()

# B) Histogram of Salary
plt.figure(figsize=(10, 6))
plt.hist(df['Salary'], bins=10, edgecolor='black')
plt.title('Histogram of Salary')
plt.xlabel('Salary')
plt.ylabel('Frequency')
plt.show()

# C) Scatter plot of Salary to Experience
plt.figure(figsize=(10, 6))
plt.scatter(df['Salary'], df['Experience'], alpha=0.7)
plt.title('Scatter Plot of Salary vs Experience')
plt.xlabel('Salary')
plt.ylabel('Experience')
plt.show()
```
