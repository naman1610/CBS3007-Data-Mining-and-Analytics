```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# Read the CSV file into a DataFrame
df = pd.read_csv('fueldata.csv')

# A) Line Chart for each fuel type
plt.figure(figsize=(10, 6))
plt.plot(df['Year'], df['Petrol Price (INR/L)'], marker='o', label='Petrol')
plt.plot(df['Year'], df['Diesel Price (INR/L)'], marker='o', label='Diesel')
plt.plot(df['Year'], df['LPG Price (INR/L)'], marker='o', label='LPG')
plt.title('Fuel Prices Over the Last 25 Years (Line Chart)')
plt.xlabel('Year')
plt.ylabel('Price (INR/L)')
plt.legend()
plt.grid(True)
plt.show()

# B) Bar Chart for each fuel type
plt.figure(figsize=(10, 6))
width = 0.25  # the width of the bars
x = df['Year']
plt.bar(x - width, df['Petrol Price (INR/L)'], width, label='Petrol')
plt.bar(x, df['Diesel Price (INR/L)'], width, label='Diesel')
plt.bar(x + width, df['LPG Price (INR/L)'], width, label='LPG')
plt.title('Fuel Prices Over the Last 25 Years (Bar Chart)')
plt.xlabel('Year')
plt.ylabel('Price (INR/L)')
plt.xticks(rotation=45)
plt.legend()
plt.show()

# C) Box Plot for each fuel type
plt.figure(figsize=(10, 6))
plt.boxplot([df['Petrol Price (INR/L)'], df['Diesel Price (INR/L)'], df['LPG Price (INR/L)']], labels=['Petrol', 'Diesel', 'LPG'])
plt.title('Fuel Prices Over the Last 25 Years (Box Plot)')
plt.ylabel('Price (INR/L)')
plt.show()

# D) Pie Chart for the distribution of fuel prices in the most recent year
recent_year = df.iloc[-1]
labels = ['Petrol', 'Diesel', 'LPG']
sizes = [recent_year['Petrol Price (INR/L)'], recent_year['Diesel Price (INR/L)'], recent_year['LPG Price (INR/L)']]
plt.figure(figsize=(8, 8))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
plt.title(f'Fuel Prices Distribution in {recent_year["Year"]} (Pie Chart)')
plt.show()

# E) Histogram for each fuel type
plt.figure(figsize=(10, 6))
plt.hist(df['Petrol Price (INR/L)'], bins=10, alpha=0.5, label='Petrol')
plt.hist(df['Diesel Price (INR/L)'], bins=10, alpha=0.5, label='Diesel')
plt.hist(df['LPG Price (INR/L)'], bins=10, alpha=0.5, label='LPG')
plt.title('Histogram of Fuel Prices')
plt.xlabel('Price (INR/L)')
plt.ylabel('Frequency')
plt.legend()
plt.show()

# F) Scatter Plot of Petrol vs Diesel Prices
plt.figure(figsize=(10, 6))
plt.scatter(df['Petrol Price (INR/L)'], df['Diesel Price (INR/L)'], alpha=0.7)
plt.title('Scatter Plot of Petrol vs Diesel Prices')
plt.xlabel('Petrol Price (INR/L)')
plt.ylabel('Diesel Price (INR/L)')
plt.grid(True)
plt.show()

# G) Heatmap of Fuel Prices
plt.figure(figsize=(10, 6))
sns.heatmap(df[['Petrol Price (INR/L)', 'Diesel Price (INR/L)', 'LPG Price (INR/L)']].corr(), annot=True, cmap='coolwarm')
plt.title('Heatmap of Fuel Prices Correlation')
plt.show()
```
